```python
# let's make a function to create 4 variables
# one filling NA with median, and another one filling NA with zeroes,mean and random values
#this way we can check which imputation can be done without distorting the distribution

def impute_na(df, variable, median,mean):
    df[variable+'_median'] = df[variable].fillna(median)
    df[variable+'_zero'] = df[variable].fillna(0) 
    df[variable+'_mean'] = df[variable].fillna(mean) 
    # random sampling
    df[variable+'_random'] = df[variable]
    # extract the random sample to fill the na
    random_sample = data[variable].dropna().sample(df[variable].isnull().sum(), random_state=0)
    # pandas needs to have the same index in order to merge datasets
    random_sample.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(), variable+'_random'] = random_sample
```

```python
mean = data.Administrative.mean()
mean
```

```python
median = data.Administrative.median()
median
```

```python
## Call impute_na function. This function creates 2 new additional variables
## this has been checked with other NA variables like Admin_duration, ProductRelated etc.
impute_na(data, 'Administrative', median,mean)
```
